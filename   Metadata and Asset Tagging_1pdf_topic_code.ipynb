```python
from PIL import Image
from PIL.ExifTags import TAGS

def extract_metadata(image_path):
    image = Image.open(image_path)
    exif_data = image.getexif()
    
    metadata = {}
    for tag_id, value in exif_data.items():
        tag = TAGS.get(tag_id, tag_id)
        metadata[tag] = value

    return metadata

# Example:
metadata = extract_metadata("example.jpg")
print("Extracted Metadata:")
print(metadata)
```

```python
asset_db = {}

def register_asset(asset_id, metadata, tags):
    asset_db[asset_id] = {
        "metadata": metadata,
        "tags": set(tags)
    }

# Example registration
register_asset(
    asset_id="file_001",
    metadata={"size": "2MB", "type": "image", "created": "2024-05-01"},
    tags=["product", "inventory", "high-res"]
)

print(asset_db)
```

```python
def add_tag(asset_id, tag):
    asset_db[asset_id]["tags"].add(tag)

def remove_tag(asset_id, tag):
    asset_db[asset_id]["tags"].discard(tag)

# Example:
add_tag("file_001", "featured")
remove_tag("file_001", "high-res")

print(asset_db["file_001"])
```

```python
def search_by_tag(tag):
    results = [
        asset_id for asset_id, data in asset_db.items()
        if tag in data["tags"]
    ]
    return results

print("Assets tagged as 'product':")
print(search_by_tag("product"))
```

```python
import json

with open("asset_metadata.json", "w") as f:
    json.dump(asset_db, f, indent=4)
```
