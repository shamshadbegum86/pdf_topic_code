````python
flowchart TD

    A[Extract Data] --> B[Validate Input]
    B --> C{Valid?}
    C -- Yes --> D[Transform & Map Attributes]
    C -- No --> E[Log Error & Skip]

    D --> F[Data Enrichment]
    F --> G[Load to Database]
    G --> H[Generate Reports]

    E --> H
````

````python
workflow = [
    ("Extract Data", "Validate Input"),
    ("Validate Input", "Valid?"),
    ("Valid?", "Transform & Map Attributes", "Yes"),
    ("Valid?", "Log Error & Skip", "No"),
    ("Transform & Map Attributes", "Data Enrichment"),
    ("Data Enrichment", "Load to Database"),
    ("Load to Database", "Generate Reports"),
    ("Log Error & Skip", "Generate Reports")
]

def generate_mermaid(workflow, output_file="workflow.md"):
    lines = ["```mermaid", "flowchart TD\n"]

    for step in workflow:
        if len(step) == 2:
            lines.append(f"    {step[0].replace(' ', '')}[{step[0]}] --> {step[1].replace(' ', '')}[{step[1]}]")
        else:
            label = step[2]
            lines.append(
                f"    {step[0].replace(' ', '')}[{step[0]}] -- {label} --> {step[1].replace(' ', '')}[{step[1]}]"
            )

    lines.append("```")

    with open(output_file, "w") as f:
        f.write("\n".join(lines))

    print(f"Workflow diagram saved to {output_file}")


generate_mermaid(workflow)
````

````python
workflow_name: customer_etl

steps:
  - name: extract
    description: Extract raw customer data from S3
    inputs: ["customer_raw.json"]

  - name: validate
    description: Apply schema checks and field validation
    validations:
      - field: age
        rule: "> 0"
      - field: email
        rule: "contains '@'"

  - name: transform
    description: Attribute mapping and standardization

  - name: enrich
    description: Add lookup values (city full name, region)

  - name: load
    description: Write enriched data to data warehouse
    outputs: ["customer_enriched.parquet"]

schedule: "0 2 * * *" # daily 2 AM
````
